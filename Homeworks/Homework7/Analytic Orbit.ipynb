```python
#Import modules
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import astropy.units as u
import astropy.constants as const
#Import other code used in this class file
from ReadFile import Read
from CenterOfMass import CenterOfMass
from GalaxyMass import ComponentsMass
```

```python
# This class uses preformatted text files to read in and read out data
# It contains several functions which calculate the force felt by M33 due to M31
# Separate accelerations are used to model an extended disk vs a bulge/halo

# These accelerations are used in a symplectic leapfrog integration to calculate the kinematics of 
# the COM of M33 which are written to the text file given as input for the class
class M33AnalticOrbit():
    
    # Initialize the class, define the variables necessary to run all functions
    def __init__(self,output):
        # output is the file name that the positions and velocities will be written to
        
        # Constants are made global for use later in the class
        self.G = (const.G.to(u.kpc**3/u.Msun/u.Gyr**2)).value
        self.fileout = str(output) + '.txt'
        
        # Initial positions and velocities are read in by the CenterOfMass class which 
        # can return COM position and velocity information
        M33 = CenterOfMass('M33_000.txt',2)
        M31 = CenterOfMass('M31_000.txt',2)
        
        # Gets the position of the COM using disk particles
        M33pos = M33.COM_P(0.1,4).value
        M31pos = M31.COM_P(0.1,4).value
        
        # Gets the velocity of the COM using it's position
        M33vel = M33.COM_V(M33pos[0],M33pos[1],M33pos[2]).value
        M31vel = M31.COM_V(M31pos[0],M31pos[1],M31pos[2]).value
        
        # Determines the separation of M33 and M31 in 3 dimensions
        self.x = (M33pos[0] - M31pos[0])
        self.y = (M33pos[1] - M31pos[1])
        self.z = (M33pos[2] - M31pos[2])
        
        # Determines the relative velocities between M33 and M31
        self.vx = (M33vel[0] - M31vel[0])
        self.vy = (M33vel[1] - M31vel[1])
        self.vz = (M33vel[2] - M31vel[2])
        
        # Scale radii and the masses of each particle type are read in or set manually
        self.rdisk = 5
        self.Mdisk = ComponentsMass('M31_000.txt',2).value * 1e12
        self.rbulge = 1
        self.Mbulge = ComponentsMass('M31_000.txt',3).value * 1e12
        self.rhalo = 52
        self.Mhalo = ComponentsMass('M31_000.txt',1).value * 1e12
        
    
    # This function calculates the acceleration of M33 due to the parts of 
    # M31 that can be modeled as a Hernquist sphere (i.e. the bulge and halo)
    def HernquistAccel(self,M,r_a,x,y,z):
        # M is the mass of either the halo or bulge of M31
        # r_a is the scale radius of the bulge or halo
        # x,y,z are the current position of the COM of M33
        
        # r is the magnitude of the separation
        r = np.sqrt(x**2 + y**2 + z**2)
        
        # Calculate the  coordinate independent portion of the acceleration
        accelmag = -1 * self.G * M / r / (r + r_a)**2
        
        # Form a vector using the magnitude of the acceleration and apply it to each coordinate
        accel = np.array([x*accelmag,y*accelmag,z*accelmag])
        
        # Returns the acceleration vector in a numpy array
        return accel
        
    # This function calculates the acceleration of M33 due to the extended disk of M31
    # The disk cannot be modeled as a Herquist sphere but rather as a torus
    def MiyamotoNagaiAccel(self,M,rd,x,y,z):
        # M is the mass of the disk
        # rd is the scale radius of the disk
        # x,y,z are the current position of the COM of M33
        
        # R is the distance between M31 and M33 on an x-y projection
        R = np.sqrt(x**2 + y**2)
        
        # B is a corrected z position to make up for the lack of spherical symmetry of the disk
        B = rd + np.sqrt(z**2 + (rd / 5.)**2)
        
        #  Calculate the  coordinate independent portion of the acceleration
        accelmag = -1 * self.G * M / ((R**2 + B**2)**1.5)
        
        # Form a vector using the magnitude of the acceleration and apply it to 
        # each coordinate with correct z correction
        accel = np.array([x*accelmag,y*accelmag,z*accelmag*B/(B - rd)])
        
        # Returns the acceleration vector as a numpy array
        return accel
        
    
    # This function sums the acceleration due to each component of M31
    def M31Acceleration(self,pos):
        # pos is a vector containing the current COM position of M33
        
        # Call earlier functions with global variables and current position
        bulgeaccel = self.HernquistAccel(self.Mbulge,self.rbulge,pos[0],pos[1],pos[2])
        diskaccel = self.MiyamotoNagaiAccel(self.Mdisk,self.rdisk,pos[0],pos[1],pos[2])
        haloaccel = self.HernquistAccel(self.Mhalo,self.rhalo,pos[0],pos[1],pos[2])
        
        # Returns a total acceleration vector as a numpy array
        return bulgeaccel + diskaccel + haloaccel
        
    # Perform a single step of integration 
    def LeapFrog(self,dt,pos0,vel0):
        # dt is the time step in Gyrs
        # pos0 is the starting position (final position of previous time step)
        # vel0 is the starting velocity (final velocity of previous time step)
        
        # Assign correct units to input values, these have been neglected previously 
        # because position and acceleration are on the same Msun-kpc-Gyr scale but velocity is in km/s
        pos0 *= u.kpc
        vel0 *= u.km / u.s
        dt *= u.Gyr
        
        # Calculate position after a half time step with inital velocity correcting for units
        poshalf = pos0 + 0.5 * dt * vel0.to(u.kpc / u.Gyr)
        
        # Calculate the acceleration felt by M33 at the half time step adding correct units
        accelhalf = self.M31Acceleration(poshalf.value) * u.kpc / u.Gyr**2
        
        # Add this acceleration to the initial velocity, again scaling to the correct units
        vel1 = (vel0 + (dt*accelhalf).to(u.km / u.s)).value
        
        # Update the position to a full time step after initial using an averaged velocity
        pos1 = (poshalf + 0.5 * dt * (vel1 * u.km / u.s).to(u.kpc / u.Gyr)).value
        # Returns two vectors, one is a position and the other a velocity both as numpy arrys
        return pos1,vel1
        
    # This integration iterates over a time range to return a separation and velocity over time
    def OrbitIntegrator(self,t0,tf,dt):
        # t0 is the initial time in Gyrs
        # tf is the end point is Gyrs
        # dt is the time step between every point in between also in Gyrs
        
        # Initialize an array to iterate over
        time = np.arange(t0,tf,dt)
        
        # Initialize an array to contain the information from the LeapFrog function
        kinematics = np.zeros((len(time),7))
        
        # Manually sets the first row giving LeapFrog a set of inital conditions
        kinematics[0] = [t0,self.x,self.y,self.z,self.vx,self.vy,self.vz]
        
        # Loop over time
        for i in range(1,len(time)):
            # Skips the inital entry to avoid overwrite
            
            # Calls LeapFrog function using previous time step results/inital conditions
            pos,vel = self.LeapFrog(dt,kinematics[i-1,1:4],kinematics[i-1,4:7])
            # Updates the contained array with current information
            kinematics[i] = [i*dt,pos[0],pos[1],pos[2],vel[0],vel[1],vel[2]]
            
        # Saves contained to a text file with proper headers
        np.savetxt(self.fileout,kinematics,header='t x y z vx vy vz',comments='#',fmt=['%.2f','%.2f','%.2f','%.2f','%.2f','%.2f','%.2f'])
        
    
```

```python
# Assign the class to a variable
Orbit = M33AnalticOrbit('M33Orbit.txt')
```

```python
# This function finds the difference between two numpy vectors
def VectorDiff(vec1,vec2):
    dif = np.asarray(vec1) - np.asarray(vec2)
    mag = np.sqrt(dif[0]**2 + dif[1]**2 + dif[2]**2)
    return mag
```
